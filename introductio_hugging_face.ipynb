## Text summarization

```python
input_text="Sure! Here’s a short paragraph The morning sun filtered gently through the leaves, casting dappled patterns on the forest floor. A cool breeze rustled the branches, carrying the fresh scent of pine and earth. Birds sang melodious tunes, their songs blending harmoniously with the soft murmur of a nearby brook. It was a serene start to the day, a moment of tranquil beauty before the world fully awoke."
print(input_text)
```

```python
import transformers
from transformers import pipeline
sumarize=pipeline("summarization")
sumarize(f"Text input:{input_text}",min_length=5,max_length=30)
```

```python
nlp=pipeline("ner")
exampl1="Obama is going Newyork"
ner_res=nlp(exampl1)
print(ner_res)
```
